```python
from collections import deque

def count_parity(puzzle):
    parity = {}
    k = 0
    for i in range(len(puzzle)):
        element = puzzle[i]
        count = 0
        for j in range(i + 1, len(puzzle)):
            next_element = puzzle[j]
            if next_element == 0:
                continue
            elif element > next_element:
                count += 1
        parity[element] = count
    print(parity)

def is_solvable(total_parity):
    if (total_parity % 2 == 0):
        print("Solvable")
    else:
        print("Unsolvable")

def print_puzzle(board):
    for row in board:
        print(" ".join(map(str, row)))
    print()

def find_blank(board):
    for i in range(3):
        for j in range(3):
            if board[i][j] == 0:
                return i, j

def is_goal(board):
    return board == [[1, 2, 3], [4, 5, 6], [7, 8, 0]]

def move_left(board):
    i, j = find_blank(board)
    if j > 0:
        board[i][j], board[i][j - 1] = board[i][j - 1], board[i][j]
        return board
    return None

def move_right(board):
    i, j = find_blank(board)
    if j < 2:
        board[i][j], board[i][j + 1] = board[i][j + 1], board[i][j]
        return board
    return None

def move_up(board):
    i, j = find_blank(board)
    if i > 0:
        board[i][j], board[i - 1][j] = board[i - 1][j], board[i][j]
        return board
    return None

def move_down(board):
    i, j = find_blank(board)
    if i < 2:
        board[i][j], board[i + 1][j] = board[i + 1][j], board[i][j]
        return board
    return None

def bfs(start):
    frontier = deque([(start, 0)])
    explored = set()

    while frontier:
        current_state, steps = frontier.popleft()

        explored.add(tuple(map(tuple, current_state)))
        for move_func in [move_left, move_right, move_up, move_down]:
            new_state = move_func([row[:] for row in current_state])
            if new_state and tuple(map(tuple, new_state)) not in explored:
                if is_goal(new_state):
                    return new_state, steps + 1
                frontier.append((new_state, steps + 1))

    return None, 0

initial_state = [
    [2, 0, 3],
    [1, 4, 5],
    [6, 7, 8]
]

parity={}

flat_list = sum(initial_state, [])

print(flat_list)

count_parity(flat_list)

total_parity = 0
for values in parity.values():
    total_parity = total_parity + values

print(f"Total parity is {total_parity}")

is_solvable(total_parity)

solution, steps = bfs(initial_state)

if solution:
    print("Solution found in", steps, "steps:")
    print_puzzle(solution)
else:
    print("No solution found.")
```
